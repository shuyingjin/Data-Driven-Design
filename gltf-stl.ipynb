```python
import os
import trimesh

# pip install trimesh


src_root = r"C:\Users\golds\Desktop\sketchfab"
dst_dir = r"C:\Users\golds\Desktop\sketchfab\converted_stl"
os.makedirs(dst_dir, exist_ok=True)


gltf_files = []
for root, dirs, files in os.walk(src_root):
    for file in files:
        if file.lower().endswith((".gltf", ".glb")):
            gltf_files.append(os.path.join(root, file))

print(f"Found {len(gltf_files)} glTF files.")


for src_path in gltf_files:
    try:

        rel_path = os.path.relpath(src_path, src_root)
        base_no_ext = os.path.splitext(rel_path)[0]
        out_filename = base_no_ext.replace(os.sep, "_") + ".stl"
        out_path = os.path.join(dst_dir, out_filename)


        mesh = trimesh.load(src_path, force="scene")
        mesh.export(out_path)
        print(f"Converted {rel_path} → {out_filename}")
    except Exception as e:
        print(f"Failed to convert {src_path}: {e}")

print(f"\nBatch conversion complete! STL files are in:\n{dst_dir}")
```
